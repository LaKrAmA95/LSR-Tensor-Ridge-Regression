### Plotting Data

```python
# Prepare data for plotting
metrics = ['NMSE', 'CORR', 'R2']
sample_labels = ['250', '350', '450', '550', '679']

# Assuming mean_df_350, mean_df_450, mean_df_550, var_df_350, var_df_450, var_df_550 are defined similarly to mean_df_250 and mean_df_679
mean_values = pd.DataFrame([mean_df_250, mean_df_350, mean_df_450, mean_df_550, mean_df_679], index=sample_labels)
var_values = pd.DataFrame([var_df_250, var_df_350, var_df_450, var_df_550, var_df_679], index=sample_labels)

import matplotlib.pyplot as plt

# Plotting
fig, axes = plt.subplots(1, 3, figsize=(10, 5), sharey=True)  # 1 row, 3 columns

for i, metric in enumerate(metrics):
    ax = axes[i]
    ax.plot(sample_labels, mean_values[metric], marker='o', linestyle='-', label=f'Mean {metric}')
    ax.fill_between(sample_labels,
                    mean_values[metric] - var_values[metric] ** 0.5,
                    mean_values[metric] + var_values[metric] ** 0.5,
                    color='gray', alpha=0.2)
    ax.set_title(f'{metric} Mean and Variance')
    ax.set_ylabel(f'{metric}')
    ax.set_xlabel('Sample Size')
    ax.legend()

plt.tight_layout()
plt.show()
```
